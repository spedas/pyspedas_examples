```python
import pyspedas
from pyspedas import tplot, options, tlimit
```

We'll load some MMS data (line plots and spectrograms) for the plot examples

```python
pyspedas.projects.mms.fpi(datatype='des-moms', trange=['2015-10-16', '2015-10-17'])
```

We'll store a list of variables to be plotted, then plot using the default settings.

```python
plot_vars = ['mms1_des_energyspectr_omni_fast', 'mms1_des_bulkv_gse_fast', 'mms1_des_numberdensity_fast']
tplot(plot_vars)
```

Zoom into a time range using the tlimit routine

```python
tlimit(['2015-10-16/11:00', '2015-10-16/12:00'])

tplot(plot_vars)
```

Reset to the full time range with the `full` option

```python
tlimit(full=True)

tplot(plot_vars)
```

You can set the y-axis range with the `options` routine.  We'll adjust the y range of the second variable being plotted.

```python
options(plot_vars[1], 'yrange', [-400, 400])

tplot(plot_vars)
```

Set the y-axis range of multiple panels

```python
options(plot_vars[0], 'yrange', [20, 1e4])
options(plot_vars[1], 'yrange', [-300, 300])
options(plot_vars[2], 'yrange', [10, 100])

tplot(plot_vars)
```

Turn logarithmic scaling on / off

```python
options(plot_vars[0], 'ylog', False)
options(plot_vars[2], 'ylog', False)

tplot(plot_vars)
```

Set the spectrogram color bar ranges using the `options` routine

```python
options(plot_vars[0], 'zrange', [1e5, 1e7])
options(plot_vars[0], 'zlog', False)

tplot(plot_vars)
```

Now we'll set some line options on the second variable.  First, some line style and width adjustments. Note how the legend changes along with the styles selected.

```python
options(plot_vars[1],'linestyle',['dot','dash','solid'])
options(plot_vars[1],'line_width',3)
tlimit(['2015-10-16/11:00', '2015-10-16/11:05'])
tplot(plot_vars[1])
```

We can change the colors using color names (as understood by Matplotlib)  We'll use orange, magenta, and cyan.

```python
options(plot_vars[1],'colors',['orange', 'magenta', 'cyan'])
tplot(plot_vars[1])
```

We can add a horizontal line at y=0 using the databar() routine.

```python
from pyspedas import databar
databar(plot_vars[1],0.0, color='black')
tplot(plot_vars[1])
```

We can add marker (symbols) in addition to the lines.

```python
options(plot_vars[1],'markers',['x','<','.'])
tplot(plot_vars[1])
```

And we can turn off the lines completely by setting the 'symbols' option to True.

```python
options(plot_vars[1],'symbols',True)
tplot(plot_vars[1])
```

We can change the location of the legend box with the 'legent_location' option.

```python
options(plot_vars[1],'legend_location','upper left')
tplot(plot_vars[1])
```

We can turn off the legend box completely by setting the 'legend_names' option to None

```python
options(plot_vars[1],'legend_names',None)
tplot(plot_vars[1])
```

We can add titles and subtitles to the X and Y axes of line plots.   The existing Y title and subtitle are fine, but let's change the xtitle and xsubtitle, and also add a panel title above.

```python
options(plot_vars[1],'xtitle','This is an X axis title.')
options(plot_vars[1],'xsubtitle', 'This is the subtitle.')
options(plot_vars[1],'title','Panel Title')
tplot(plot_vars[1])
```

We can change the color of the x, y, and (for spectrograms) z titles.  Subtitles will inherit the properties of the titles.
The font sizes for x/y/z/panel titles and subtitles share a single option 'char_size'.

```python
options(plot_vars[1],'xtitle_color','blue')
options(plot_vars[1],'ytitle_color','green')
options(plot_vars[1],'char_size',20)
tplot(plot_vars[1])
```

We can save plot output to a file, with several format option, using tplot keyword arguments save_png, save_jpeg, save_eps, save_svg, or save_pdf, and suppying a filename.  We'll zoom back out again, and display all three panels.



```python
tlimit(full=True)
tplot(plot_vars,save_png='saved_example_plot.png')
```

We can return the Matplotlib figure and axes objects by setting the tplot keyword 'return_plot_objects' to True.
This is useful if you want to modify the plot using Matplotlib directly, or if you want to do an interactive time
selection with the ctime() utility.

```python
fig, axes = tplot(plot_vars,return_plot_objects=True)
help(fig)
```
