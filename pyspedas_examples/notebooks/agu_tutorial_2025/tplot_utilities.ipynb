## Load some data

We'll use MMS FPI ion distribution data as an example; this is a good case to show because the data are multi-dimensional

```python
import pyspedas

pyspedas.projects.mms.fpi(trange=['2015-10-16/13:05', '2015-10-16/13:06'],
                 data_rate='brst',
                 datatype=['dis-dist', 'dis-moms'])
```

## Time series data

The data are now loaded into `tplot` variables (the `t` stands for `timeseries`!); access the data:

```python
from pyspedas import get_data
```

```python
data = get_data('mms1_dis_dist_brst', dt=True)
```

get_data() returns a Python tuple with named fields:

data.times:  Unix timestamps

data.y: A one-dimensional or multidimensional array of data values

data.v (and possibly v1, v2, v3):  For multi-dimensional data, these values may represent coordinate values along
each axis of the data array.  For example, an energy spectrogram might have a 'v' attribute representing the center energies of each bin.  For constant bin values, v might be a 1-D array of N energies (where N is the first non-time dimension of the data array).   Time-varying bin values are also supported, where 'v' can be a 2-D array of ntimes x nbins.

```python
data.times
```

## Store metadata alongside actual data

You can also access the metadata of a tplot variable using `get_data`:

```python
metadata = get_data('mms1_dis_dist_brst', metadata=True)
```

This is just a dictionary, containing the CDF and plot metadata:

```python
metadata.keys()
```

So to get the CDF metadata for our DF variable:

```python
metadata['CDF']['VATT']
```

If the CDF metadata includes a coordinate system (which this one doesn't), it'll be available using `pyspedas.get_coords`

```python
from pyspedas import get_coords
```

```python
get_coords('mms1_dis_dist_brst')
```

You can set it with `pyspedas.set_coords`

```python
from pyspedas import set_coords
```

```python
set_coords('mms1_dis_dist_brst', 'DBCS')
```

And now it's available:

```python
get_coords('mms1_dis_dist_brst')
```

## Combine multiple timeseries objects, and keep track of metadata

You can split a tplot variable into individual components using `split_vec`, and combine them using `join_vec`, e.g., 

Note: keeping track of the metadata could be improved

```python
from pyspedas import join_vec, split_vec
```

```python
split_vec('mms1_edp_dce_gse_brst_l2')
```

Plot the individual components:

```python
from pyspedas import tplot

tplot(['mms1_edp_dce_gse_brst_l2_x',
       'mms1_edp_dce_gse_brst_l2_y',
       'mms1_edp_dce_gse_brst_l2_z'])
```

And you can use `join_vec` to join them back together:

```python
join_vec(['mms1_edp_dce_gse_brst_l2_x',
          'mms1_edp_dce_gse_brst_l2_y',
          'mms1_edp_dce_gse_brst_l2_z'], newname='efield_from_components')
```

```python
tplot('efield_from_components')
```
