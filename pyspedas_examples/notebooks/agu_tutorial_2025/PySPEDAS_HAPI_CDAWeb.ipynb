# Loading data via CDAWeb

SPDF offers a web service, CDAWeb,for accessing its archive holdings.  Like HAPI, it offers ways to query for missions
and data sets, and to download a selected data set for use in PySPEDAS

Import PySPEDAS CDAWeb interface and a few other tools, then get a catalog of available observatories

```python
from pyspedas import CDAWeb, data_exists, del_data, get_data, time_double
from pyspedas.cdagui_tools.config import CONFIG

# This is an object oriented interface.  We'll create a CDAWeb object, then call various methods on it.

cdaweb_obj = CDAWeb()
obslist = cdaweb_obj.get_observatories()
print("First five observatories returned...", obslist[0:5])
```

Now get the list of instrument types supported

```python
inst_list = cdaweb_obj.get_instruments()
print("First five instrument type lists returned...", inst_list[0:5])

```

Get a list of datasets available for Voyager magnetic field measurements

```python
ds_list = cdaweb_obj.get_datasets(["Voyager"], ["Magnetic Fields (space)"])
print(ds_list)
```

Using a dataset from the above list, and a time range, return a list of download URLs

```python
urllist = cdaweb_obj.get_filenames(
            ["VOYAGER2_2S_MAG (1977-08-24 to 1991-01-01)"],
            "1979-01-11",
            "1979-01-12"
        )
print(urllist)
```

Now we'll request a year's worth of data from a different data set, and plot it.

```python
# We'll pick one of available data sets and load it into tplot variables
dataset = "VOYAGER2_COHO1HR_MERGED_MAG_PLASMA"
start_time = "2009-01-01 00:00:00"
end_time = "2010-01-01 00:00:00"
output_dir = CONFIG["local_data_dir"]

# Get the URLs for the available data in this time range
urllist = cdaweb_obj.get_filenames([dataset], start_time, end_time)

# Download the data and load as tplot variables.  Setting a prefix
# is useful if you want to work with both Voyager 1 and Voyager 2
# data; the variable names in the archived data are the same for both
# spacecraft.

cdaweb_obj.cda_download(urllist, output_dir, prefix="v2_")
v2_vars=pyspedas.tnames('v2_*')
print(v2_vars)
tplot('v2_protonDensity')
```
